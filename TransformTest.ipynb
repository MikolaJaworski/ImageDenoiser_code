```python
def rotate(img, rang):
    rotate=iaa.Affine(rotate=rang)
    rotate_img=rotate.augment_image(img)
    return rotate_img

def shear(img, rang):
    shear = iaa.Affine(shear=rang)
    shear_img = shear.augment_image(img)
    return shear_img

def crop(img, rang):
    crop = iaa.Crop(percent=rang)
    crop_img=crop.augment_image(img)
    return crop_img

def fliphr(img):
    flip_hr = iaa.Fliplr(p=1.0)
    flip_img = flip_hr.augment_image(img)
    return flip_img

def flipvr(img):
    flip_vr = iaa.Flipud(p=1.0)
    flip_img = flip_vr.augment_image(img)
    return flip_img

def contr(img, gamma):
    contr = iaa.GammaContrast(gamma=gamma)
    contr_img = contr.augment_image(img)
    return contr_img

def scale(img, x, y):
    scale_im=iaa.Affine(scale={"x": x, "y": y})
    scale_img = scale_im.augment_image(img)
    return scale_img

def save_agu_img(path, img, img_name, agum):
    img_name = '{}_{}.jpg'.format(img_name, agum)
    imageio.imwrite(os.path.join(path, img_name), img)    
```

```python
save_path = './BSDS300-images/BSDS300/images/train'
r_rang = (-60, 60)
s_rang = (0, 40)
c_rang = (0, 0.3)
gamma = 1.5
x = (1.5, 1.0)
y = (1.5, 1.0)
```

```python
for i in range(len(images)):
    save_agu_img(save_path, crop(images[i], c_rang), i, 'crop')
```

```python
for i in range(len(images)):    
    save_agu_img(save_path, rotate(images[i], r_rang), i, 'rotate')
```

```python
for i in range(len(images)):    
    save_agu_img(save_path, shear(images[i], s_rang), i, 'shear')
```

```python
for i in range(len(images)):   
    save_agu_img(save_path, fliphr(images[i]), i, 'fliphr')
```

```python
for i in range(len(images)):    
    save_agu_img(save_path, flipvr(images[i]), i, 'flipvr')
```

```python
for i in range(len(images)):    
    save_agu_img(save_path, contr(images[i], gamma), i, 'contrast')
```

```python
for i in range(len(images)):    
    save_agu_img(save_path, scale(images[i], x, y), i, 'scale')
```
